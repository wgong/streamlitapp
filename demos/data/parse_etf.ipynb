```python
l = [t.strip() for t in txt.split("\n") if t.strip()]
```

```python
data = []
for i in l:
    t2 = i.split(";")
    order = t2[-1].strip()
    sym = t2[0].split()[0]
    name = " ".join(t2[0].strip().split()[1:])
    data.append([sym, name, "International", 0, order])
```

```python
data
```

```python
s = ""
for d in data:
    s += ",".join([str(i) if type(i) != str else i for i in d])
    s += "\n"
```

```python
s
```

```python
open("intl-etf.csv", "w").write(s)
```
