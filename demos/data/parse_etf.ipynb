```python
import pandas as pd

df = pd.read_csv("wl_futures_etf.csv")
```

```python
df
```

```python
sectors = df["sector"].unique().tolist()
```

```python
sectors
```

```python
etf_dict = {}
for sect in sectors:
    sym_name = df[df["sector"] == sect][["symbol","name"]]
    etf_dict[sect] = dict(zip(sym_name.symbol, sym_name.name))
```
